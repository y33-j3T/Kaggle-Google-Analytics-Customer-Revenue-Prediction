### Drop constant features

```python
#constant features of train_flattened
constF_train = [col for col in train_flattened.columns if train_flattened[col].nunique(dropna=False)==1 ]
constF_train
```

```python
#constant features of test_flattened
constF_test = [c for c in test_flattened.columns if test_flattened[c].nunique(dropna=False)==1 ]
constF_test
```

```python
#drop constant features of train_flattened
train_flattened_clean = train_flattened.drop(columns=constF_train)
train_flattened_clean.head()
```

```python
#drop constant features of test_flattened
test_flattened_clean = test_flattened.drop(columns=constF_test)
test_flattened_clean.head()
```
